### Missing Values

```python
#features_with_n is a list containing nan values
features_with_n = [features for features in dataset.columns if dataset[features].isnull().sum()>1]

#iterating through all the features and finding the mean of feature and printing it
for i in features_with_n:
    print(i, np.round(dataset[i].isnull().mean() , 4), '% missing values' )
```

 ### Since there are many missing values, we need to find the relationship between the missing values and Sales Price
Let's plot some diagram for this

```python
palette = sns.color_palette("Set2", 4)
for feature in features_with_n:
    data = dataset.copy()
    # let's make a variable that indicates 1 if the observation was missing else it have it's value as 0
    data[feature] = np.where(data[feature].isnull(), 1,0)
    #Let's calculate the mean saleprice where the information is missing or present
    data.groupby(feature)['SalePrice'].median().plot.bar(color=palette)
    plt.title(feature)
    plt.show()
```

Here with the relation between the missing values and the dependent variable is clearly visible. So We need to replace these nan values with something meaningful which we will do in the Feature Engineering Section.

### Numerical Variable

```python
# how many features are numerical variable
numerical_features = [feature for feature in dataset.columns if dataset[feature].dtype != 'O']   # O--> Object(string , by default it will be numerical)
print("Number of numerical Variables:" , len(numerical_features))

#visualise the numerical variable
dataset[numerical_features].head()
```

### Temporal Variables (Eg : Datetime Variables)

From the Dataset we have 4 year variables. We have to extract information from the datetime variables like no of years or no. of days. One example in this specific scenario can be difference in years between the year the house was built and the year the house was sold. We will be performing this analysis is the Feature Engineering.

```python
# year_feature - is a list of variables that contain year information
year_feature = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]
year_feature
```

```python
#Exploring the content of these Variables
for feature in year_feature:
    print(feature, dataset[feature].unique())
```

```python
# Let's analyze the temporal datetime variables
# is there any relation between year the house was sold and sales price

dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title("House Price VS Pear Sold")
```

```python
year_feature
```

```python
# We will compare the difference between all year feature with sales price

for feature in year_feature:
    if feature!='YrSold':
        data=dataset.copy()
        data[feature] = data['YrSold'] - data[feature]
        plt.scatter(data[feature], data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

```python
# numerical variables are usually of 2 types
# Continous variable and discrete variable

discrete_feature=[feature for feature in numerical_features if len(dataset[feature].unique())<25 and feature not in year_feature+['Id']]
print("Discrete Variables Count: {}".format(len(discrete_feature)))
```

```python
discrete_feature
```

```python
dataset[discrete_feature].head()
```

```python
## Let's find the relationship between and sale Price
palette = sns.color_palette("Set2", 15)
for feature in discrete_feature:
    data=dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar(color=palette)
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

```python
# There is relationship between variable number and SalePrice
```

```python
continuous_feature=[feature for feature in numerical_features if feature not in discrete_feature+year_feature+['Id']]
print("Continuous feature Count {}".format(len(continuous_feature)))
```

```python
## Lets analyse the continuous values by creating histograms to understand the distribution

for feature in continuous_feature:
    data=dataset.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    plt.show()
```

## Exploratory Data Analysis Part - 2

```python
# logarithmic transformation

for feature in continuous_feature:
    data = dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        data['SalePrice'] = np.log(data['SalePrice'])
        plt.scatter(data[feature] , data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.title(feature)
        plt.show()
```

### Outliers
